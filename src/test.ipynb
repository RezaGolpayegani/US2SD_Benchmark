```python
import csv
from collections import Counter
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open("../data/All_US.csv", mode='r', newline='') as file:
    reader = csv.reader(file)
    my_list = []
    for row in reader:
        if row[3] == 'Project Name':
            continue
        my_list.append(row[3])
```

```python
data = Counter(my_list)
```

```python
project_name = list(data.keys())
number_of_USs = list(data.values())
```

```python
df = pd.read_csv("../data/All_US.csv")
```

```python
sns.set_style("darkgrid")
sns.set_palette(["#FF5733", "#008080", "#800080"])

sns.catplot(
    y = 'Project Name',
    kind = 'count',
    data = df,
    height = 6,
    aspect = 1.5)
plt.title("Data Distribution ")
plt.tight_layout()
plt.savefig("../images/Data_Dist.png")
```
